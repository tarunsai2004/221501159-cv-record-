```python
import cv2
from google.colab.patches import cv2_imshow  # For displaying images in Google Colab

# Load the pre-trained face detection model (Haar Cascade)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image from file
image_path = '/content/vk.jpg'  # Replace with your image path
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print("Failed to load image. Please check the file path.")
else:
    # Convert the image to grayscale for better face detection
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the grayscale image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around the detected faces and add smaller labels below the rectangle
    for i, (x, y, w, h) in enumerate(faces):
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        label = f'Person {i + 1}'
        # Put the label *below* the rectangle, and use a smaller font size
        cv2.putText(image, label, (x, y + h + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Display the image with the detected faces and labels in Colab
    cv2_imshow(image)

    # # Wait until the user manually closes the image window (useful for Colab)
    # cv2.waitKey(0)
```
